```python

    
all_stats = pd.DataFrame({}) 
if with_FL:
    #m_values = np.linspace(0,1,20)
    b_values = np.linspace(0.01,1,7)
    b_values = np.linspace(0.01,1,3)

    m_values = np.linspace(0.01,0.3,7)
    d_values = np.linspace(0.01,0.7,7)
    f_values = np.linspace(0.1,3,7)
    c_values = np.linspace(0.01,1,7)
    r_values = np.linspace(0.001,0.3,7)
    delta_values = np.linspace(0.1,1,7)
    alpha_f_values = np.linspace(0.1,2,7)
    alpha_b_values = np.linspace(0.1,2,3)


else:
    b_values = np.linspace(0,1,10)
    m_values = np.linspace(0,1,10)
    d_values = np.linspace(0,1,10)
    

m=0.2
b=0.55
f =  params["f"]
d = params["d"]
c = params["c"]
r = params["r"]
delta = params["delta"]
alpha_f = params["alpha_feedback"]

# change these 2 lines
param_to_vary = "b"    
param_values = b_values
replications = 1
# change this line 
for b in param_values:
    print(param_to_vary ,alpha_b)
    for _ in range(replications):
        model = EcoModel(b, m, d,f,c,r,delta, alpha_f, alpha_b, config_file)
        for i in range(70):
            model.step()
        stats = model.datacollector.get_model_vars_dataframe()
        stats_last = stats.tail(1)
        # Change this line
        stats_last[param_to_vary] = b
        all_stats = all_stats.append(stats_last)
all_stats['total'] = (all_stats.Degraded+all_stats.Vegetated+all_stats.Empty)

print(time.time()-start_time)
all_stats
```
