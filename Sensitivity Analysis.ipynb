```python
output_dir = "output/"
# run.py
from mesa.batchrunner import BatchRunner

from ecosystem.model import EcoModel
```

```python
import time
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

start_time = time.time()

delta = 0
c = 0.3
r = 0
d = 0.2
f = 0.8
m = 0.07
b = 0.4
# fixed_params = {"width": 50,
#                 "height": 50,
#                 "b": 0.5,
#                 "m": 0.1}


#def evaluate_model(w,h,m,b):
#    model = EcoModel(50, 50, b, 0.07)
#    stats = model.datacollector.get_model_vars_dataframe()
#    stats_last = stats.tail(1)
#    return(stats_last.Vegetated)

    
all_stats = pd.DataFrame({}) 
#m_values = np.linspace(0,1,20)
b_values = np.linspace(0,1,10)
m_values = np.linspace(0,0.1,10)
d_values = np.linspace(0,1,10)
# change these 3 lines
param_to_vary = "b"    
param_values = b_values

replications = 5
for b in param_values:
    for _ in range(replications):
        model = EcoModel(50, 50, b, m, d)
        for i in range(150):
            model.step()
        stats = model.datacollector.get_model_vars_dataframe()
        stats_last = stats.tail(1)
        # Change
        stats_last[param_to_vary] = b
        all_stats = all_stats.append(stats_last)
all_stats['total'] = (all_stats.Degraded+all_stats.Vegetated+all_stats.Empty)

print(time.time()-start_time)
all_stats
```
